## Creating Agents

- Define your Agents, and provide them a `role`, `goal` and `backstory`.
- It has been seen that LLMs perform better when they are role playing.

### Agent: Planner

```python
planner = Agent(
    role=" Content Planner",
    goal="Plan engaging, factually accurate, and comprehensive, technical content on {topic} for a machine learning related blog by gathering information from the given YouTube video",
    backstory=
              "You're an expert content planner, working on planning a comprehensive technical article about the topic: {topic} for a machine learning blog with 1000+ subscribers. "
              "You collect information and prepare an outline that helps the audience learn about the {topic} in detail by researching the given YouTube Video. "
              "Your work is the basis for the Content Writer to write a comprehensive article on this topic.",
    allow_delegation=False,
    tools = [yt_tool],
	verbose=True
)
```

### Agent: Writer

```python
writer = Agent(
    role="Content Writer",
    goal="Write insightful and factually accurate "
         "comprehensive, technical blog post about the topic: {topic} ",
    backstory="You're an expert content writer, working on a writing a technical blog post about the topic: {topic} for a machine learning blog with 1000+ subscribers. "
              "You base your writing on the work of the Content Planner, who provides an outline about the topic: {topic} as well as the YouTube video. "
              "Ensure that your article includes the following sections."
              "1. Non-Agentic versus Agentic Workflow\n"
              "2. Performance comparison against benchmarks\n"    
              "3.Agentic Reasoning Design Patterns\n"    
              "3a.Reflection\n"  
              "3b.Tool Use\n"      
              "3c.Planning\n"  
              "3d.Multi-Agent Collaboration\n"      
              "4.Conclusion\n",
    allow_delegation=False,
    tools = [yt_tool],
    verbose=True
)
```

### Agent: Editor

```python
editor = Agent(
    role="Content writer",
    goal="Edit a given blog post to align with "
         "the recommended writing style. ",
    backstory="You are a content editor working for a machine learning blog post."
              "You need to ensure that the final article follows blog writing best practices detailed below. "
              "1. Use less than 30 words per sentence.\n"
              "2. Eliminate weasel words.\n"
              "3. Use active voice.\n"
              "4. Avoid cluttered words and buzzwords.\n"
              "5. Spell out the acronyms upon its first occurence. Avoid jargons.\n"
              "6. Ensure there is a flow to the story with beginning, middle and end.\n",
    allow_delegation=False,
    verbose=True
)
```

## Creating Tasks

- Define your Tasks, and provide them a `description`, `expected_output` and `agent`.

### Task: Plan

```python
plan = Task(
    description=(
        "Research and collect detailed info from the YouTube video on the {topic} to do the below"
        "1. Prioritize telling a good story with a clear, easy to understand and compelling flow on the {topic}.\n"
        "2. Get detailed information on the {topic}.\n "
    ),
    expected_output="A comprehensive content plan document "
        "with a well-defined outline and well researched content based on the {topic} of YouTube video.",
    agent=planner,
)
```

### Task: Write

```python
write = Task(
    description=(
        "1. Use the info from the Content planner and the YouTube video on the {topic} to craft a compelling "
            "comprehensive, blog post on {topic}.\n"
        "2. Provide as much detail as possible."
    ),
    expected_output="A well-written comprehensive blog post that is 1600 words long"
        "in markdown format, ready for publication. ",
    agent=writer,
)
```

### Task: Edit

```python
edit = Task(
    description=("Proofread the given blog post for "
                 "grammatical errors and "
                 "alignment with the recommended writing best practices as detailed by the editor agent."),
    expected_output="A well-written comprehensive, blog post in markdown format, "
                    "ready for publication. ",
    agent=editor
)
```

## Creating the Crew

- Create your crew of Agents
- Pass the tasks to be performed by those agents.
    - **Note**: *For this simple example*, the tasks will be performed sequentially (i.e they are dependent on each other), so the _order_ of the task in the list _matters_.
- `verbose=2` allows you to see all the logs of the execution. 

```python
crew = Crew(
    agents=[planner, writer, editor],
    tasks=[plan, write, edit],
    memory = True,
    cache = True,
    verbose=2
)
```

## Running the Crew

**Note**: LLMs can provide different outputs for they same input, so what you get might be different than what you see in the video.

```python
result = crew.kickoff(inputs={"topic": "AI agentic workflows"})
```
